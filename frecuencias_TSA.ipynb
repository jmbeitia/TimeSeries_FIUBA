```python
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
archivo_json = 'mediciones_filtrado.json'

with open(archivo_json, 'r') as archivo:
    datos_json = json.load(archivo)
```

```python
df_filtrado = pd.DataFrame.from_dict(datos_json)
df_filtrado['Timestamp'] = pd.to_datetime(df_filtrado['Timestamp'], unit='ms')
df_filtrado
```

```python
df_filtrado.to_csv('datasets/frecuencia.csv')
```

```python
df_filtrado.dtypes
```

```python
df_filtrado['Frecuencia'].describe()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_filtrado['Timestamp'], df_filtrado['Frecuencia'], label='Frecuencia')
plt.title('Serie de Tiempo Completa de Frecuencia')
plt.xlabel('Fecha')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
datos_del_dia = df_filtrado[(df_filtrado['Timestamp'] >= '2023-12-15 00:00:00') & (df_filtrado['Timestamp'] <= '2023-12-15 23:59:59')]

# Graficar la serie de tiempo
plt.figure(figsize=(10, 6))
plt.plot(datos_del_dia['Timestamp'], datos_del_dia['Frecuencia'])
plt.title('Serie de Tiempo de la Frecuencia el 15 de Diciembre de 2023')
plt.xlabel('Hora del dÃ­a')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
datos_por_hora = df_filtrado[(df_filtrado['Timestamp'] >= '2023-12-15 00:00:00') & (df_filtrado['Timestamp'] <= '2023-12-15 00:59:59')]

plt.figure(figsize=(10, 6))
plt.plot(datos_por_hora['Timestamp'], datos_por_hora['Frecuencia'])
plt.title('Frecuencia de la primera hora del 15 de Diciembre de 2023')
plt.xlabel('Hora del dÃ­a')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
datos_por_10_min = df_filtrado[(df_filtrado['Timestamp'] >= '2023-12-15 00:00:00') & (df_filtrado['Timestamp'] <= '2023-12-15 00:09:59')]

plt.figure(figsize=(10, 6))
plt.plot(datos_por_10_min['Timestamp'], datos_por_10_min['Frecuencia'])
plt.title('Frecuencia de los primeros 10 minutos el 15 de Diciembre de 2023')
plt.xlabel('Hora del dÃ­a')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
