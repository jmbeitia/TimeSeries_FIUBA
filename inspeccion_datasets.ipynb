```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración para visualizaciones
%matplotlib inline
sns.set_theme(style="darkgrid")
```

```python
catfish_df = pd.read_csv('datasets/catfish.csv', parse_dates=['Date'])
rigo_df = pd.read_csv('datasets/RIGO.2012.2021.csv', parse_dates=['fechaHora'])
ypfd_df = pd.read_csv('datasets/YPFD.2000.2021.csv', parse_dates=['fechaHora'])
```

```python
# Catfish
print("Estadísticas Descriptivas - Catfish:")
print(catfish_df.describe())

# RIGO
print("\nEstadísticas Descriptivas - RIGO:")
print(rigo_df['ultimoPrecio'].describe())

# YPFD
print("\nEstadísticas Descriptivas - YPFD:")
print(ypfd_df['ultimoPrecio'].describe())
```

# Catfish


```python
plt.figure(figsize=(10, 6))
plt.plot(catfish_df['Date'], catfish_df['Total'], label='Total por mes')
plt.title('Producción Mensual de Catfish')
plt.xlabel('Fecha')
plt.ylabel('Total')
plt.legend()
plt.show()
```

# RIGO

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=rigo_df, x='fechaHora', y='ultimoPrecio').set_title('Precio Último de RIGO a lo largo del tiempo')
plt.xlabel('Fecha')
plt.ylabel('Último Precio')
plt.show()
```

# YPFD

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=ypfd_df, x='fechaHora', y='ultimoPrecio').set_title('Precio Último de YPFD a lo largo del tiempo')
plt.xlabel('Fecha')
plt.ylabel('Último Precio')
plt.show()
```
